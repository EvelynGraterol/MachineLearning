```python
#Importando librerias
import numpy  as np
import matplotlib.pyplot as plt
```

### 1.  Grafico a partir de lista : x--> indice del vector y-->vector

```python
ahorros = [50, 100, 30, 65]
plt.plot(ahorros)
```

### 2. Graficos a partir de un arrays de numpy

```python
ahorros = np.random.randint(100,size=[12]) # size, se refiere al máximo valor de x

plt.plot(ahorros)
plt.show()
```

### 3.Mapeado de los textos en los ejes

```python
#Por defecto en los graficos no podemos mostrar texto en los ejes
#solucion:mapear los textos en una lista del mismo tamano que el numero de los valores del eje x

ahorros = np.random.randint(100, size=[6])
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio']
mapeo = range(len(meses))


plt.plot(ahorros)
plt.xticks(mapeo, meses)
plt.show()
```

```python
#Nota para concluir: Mientras nos encontremos en graficos de una dimension, el metodo .plot() toma los valores del eje Y y asigna 
#valores a X automaticamente, a partir de la logitud del vector.
```

### 4.Definiendo el vector X e Y

```python
x = np.arange(6)
y = np.random.randint(20, size = [6])
plt.plot(x, y)
plt.show()
```

### 5.Cambio de ejes

```python
plt.plot(y, x)
plt.show()
```

### 6.Haciendo aleatorio ambos ejes

```python
x = np.random.randint(20, size = [6])
y = np.random.randint(20, size = [6])

plt.plot(x, y)
plt.show()

###Nota : Al generarar valores aleatorios para x, y se rompe la lógica. Porque los gráficos se rigen en función del 
### tiempo y eso implica que uno de los ejes debe ser un valor incremental.
```

### 7.Trabajando con los limites del grafico.

```python
ahorros = np.random.randint(100, size=[6])
meses = ['Enero', 'Febrero','Marzo','Abril','Mayo', 'Junio']
mapeo = range(len(meses))

plt.plot(ahorros)
plt.xticks(mapeo, meses) 
plt.ylim(0, 100)           #Escala para el eje y
plt.show()
```

```python
#Utilizando los limites podemos centrarnos en cualquier parte del gráfico
plt.plot(ahorros)
plt.xticks(mapeo, meses)
plt.xlim(2, 4)
plt.ylim(0, 100)
plt.show()
```

```python
plt.plot(ahorros)
plt.xticks(mapeo, meses)
plt.xlim(2, 4)    #plt.xlim y plt.ylim se comportan como 
plt.ylim(40, 60)  #zoom dentro de la grafica.
plt.show()
```

```python
#Titulo,leyendas y etiquetas

ahorros = np.random.randint(100, size=[6])
meses = ['Enero', 'Febrero','Marzo','Abril','Mayo', 'Junio']
mapeo = range(len(meses))

plt.plot(ahorros)
plt.xticks(mapeo, meses)
plt.title('Ahorros')
plt.xlabel('Meses')
plt.ylabel('Dinero en $')
plt.show()
```

```python
plt.plot(ahorros, label='Evolución')
plt.xticks(mapeo, meses)
plt.xlim(2, 4)
plt.title('Ahorros')
plt.xlabel('Meses')
plt.ylabel('$')
plt.legend(loc=4)
plt.show()
```
