```python
X = df.drop(columns=['Yearly Amount Spent'], axis=1)
y = df['Yearly Amount Spent']
```

###### Aplicación de standard scaler

```python
col_names = ['Avg. Session Length', 'Time on App', 'Time on Website', 'Length of Membership']
```

```python
scaler = StandardScaler().fit(X.values)
features = scaler.transform(X.values)
X = pd.DataFrame(features, columns = col_names)
```

### MODELO: SVR Support vector regressor -> Varible a predecir- continua

###### Entrenamiento del modelo

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=41)
```

```python
svr_p = SVR(kernel='linear', C=10, cache_size=700,verbose=True) 
```

```python
#Entrenar
svr_p.fit(X_train, y_train)
```

```python
y_pred = svr_p.predict(X_test)
```

```python
y_pred_train = svr_p.predict(X_train)
```

### Verificacion de los resultados 

###### SIN SCALER!

###### Errores del train

```python
print('MAE', mean_absolute_error(y_pred_train, y_train))
```

```python
print('RMSE', np.sqrt(mean_squared_error(y_pred_train, y_train)))
```

###### Errores del test

```python
print('MAE:', mean_absolute_error(y_pred, y_test))
```

```python
print('RMSE:', np.sqrt(mean_squared_error(y_pred, y_test)))
```

###### Coecificientes del regresor

```python
pd.DataFrame(svr_p.coef_, columns=X_train.columns).T
```

###### CON SCALER!

```python
print('MAE', mean_absolute_error(y_pred_train, y_train))
```

```python
print('RMSE', np.sqrt(mean_squared_error(y_pred_train, y_train)))
```

```python
print('MAE:', mean_absolute_error(y_pred, y_test))
```

```python
print('RMSE:', np.sqrt(mean_squared_error(y_pred, y_test)))
```
