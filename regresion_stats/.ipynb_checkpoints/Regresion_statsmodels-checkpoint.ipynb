### Regresión lineal (Mínimos cuadrados ordinarios OLS)

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from patsy import dmatrices #Llamado la libreria patsy que se comporta como R

pd.options.display.max_columns = None
```

```python
history_document = sm.datasets.get_rdataset("Guerry", "HistData")
```

```python
#Documentacion del dataset que se esta empleando
print(history_document.__doc__)
```

```python
#carga del dataset
df = sm.datasets.get_rdataset("Guerry", "HistData").data
```

```python
df.head()
```

```python
df.columns
```

```python
#Seleccion de variables a trabajar
col = ['Department','Literacy','Wealth','Lottery','Region']
```

```python
df = df[col]
```

```python
df.head()
```

```python
#construccion para la regresion lineal
y, X = dmatrices('Lottery ~ Literacy + Wealth + Region', data = df, return_type='dataframe')
```

```python
#Fit del modelo

mod = sm.OLS(y, X)

res = mod.fit()

print(res.summary())
```
